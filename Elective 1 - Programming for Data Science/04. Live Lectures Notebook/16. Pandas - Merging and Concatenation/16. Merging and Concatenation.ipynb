# Merging

```python
import numpy as np
import pandas as pd
```

## Sample Data

```python
data_1 = {'Student': ['Pedro','Jose','Petra','Sam','Billy','Melandra','Johnny'], 
          'Section': ['A','B','A','B','C','C','C'],
          'Grades':  ['82','85','90','89','92','90','83']
}
```

```python
data_2 = {'Student': ['Jose','Pedro','Sam','Billy','Petra','Melandra'],
          'School' : ['School 1','School 1','School 2','School 3', 'School 2','School 3']
          
}
```

```python
df_1 = pd.DataFrame(data_1)
df_1
```

```python
df_2 = pd.DataFrame(data_2)
df_2
```

## Merging Data Frame

```python
# Inner

merge_df1 = pd.merge(df_1, df_2, how = 'inner', on = 'Student')
merge_df1
```

```python
# Left

pd.merge(df_1, df_2, how = 'left', on = 'Student')
```

```python
# Right

pd.merge(df_1, df_2, how = 'right', on = 'Student')
```

```python
# Outer

pd.merge(df_1, df_2, how = 'outer', on = 'Student')
```

## Concatenation

```python
df_3 = pd.DataFrame({'Residence': ['Laguna','Quezon','Cavite','Batangas','Rizal','NCR','Batangas']})
df_3
```

```python
df_concat = pd.concat([df_1,df_3], axis = 1, sort = True)
df_concat
```

```python
df_concat = pd.concat([df_1, df_2, df_3], axis = 1)
df_concat
```
