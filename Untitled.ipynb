# Data Visualization (Pandas)

```python
# Data Visualization
# 1. Pandas
# 2. MatplotLib
# 3. Seaborn
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Data

df = pd.DataFrame(np.random.randn(500), index = pd.date_range('2019-01-01', periods = 500), columns = ['values'])
df
```

### Line Plot

```python
df.plot()
```

```python
# Figure Size

df.plot(figsize = 15,8)) # tuple
```

```python
# Another data set (Iris Dataset)

iris = sns.load_dataset('iris')
iris
```

```python
iris.plot(figsize = (10,5), title = 'Iris Dataset')
```

## Data Visualization (Pandas - Bar Plot)

```python
c_data = iris.drop(['species'], axis = 1)
c_data
```

```python
c_data.iloc[3].plot(kind = 'bar', title = 'First Sample')
```

```python
c_data.iloc[121].plot(kind = 'barh', title = 'First Sample')
```

## Data Visualization (Pandas - Stacked Plot)

```python
# Data
np.random.seed(10)

df2 = pd.DataFrame(np.random.randn(10,4), columns = ['a','b','c','d'])
df2
```

```python
df2.plot(kind = 'bar', stacked = True, title = 'Samples')
```

## Data Visualization (Histogram)

- Distribution of Data

```python
# Data

c_data
```

```python
c_data.plot(kind = 'hist', title = 'Iris dataset', bins = 20)
```

```python
c_data.plot(kind = 'hist', title = 'Iris dataset', bins = 20, stacked = True)
```

```python
c_data.plot(kind = 'hist', title = 'Iris dataset', bins = 20, stacked = True, orientation = 'horizontal')
```

```python
# Alternate Histogram (Additional knowledge)

c_data.hist(color = 'r')
```

# Data Visualization (Box Plot)

```python
c_data
```

```python
c_data.plot(kind = 'box', figsize = (16, 6))
```

```python
c_data.describe()
```

```python
# Orientation

c_data.plot(kind = 'box', figsize = (16, 6), vert = False)
```

# Data Visualization (Area Plot)

```python
c_data
```

```python
c_data.plot(kind = 'area', figsize = (10,5), title = 'Area Graph')
```

```python
# Non-Stacked, Transparency

c_data.plot(kind = 'area', figsize = (10,5), title = 'Area Graph', stacked = False, alpha = 0.25)
```

# Data Visualization (Scatter Plot)

```python
c_data
```

```python
# x and y axis values

c_data.plot(x = 'sepal_width', y = 'petal_width', kind = 'scatter', title = 'Iris Data')
```

```python
# with heat map

c_data.plot(x = 'sepal_width', y = 'petal_width', kind = 'scatter', title = 'Iris Data', c = 'petal_width')
```

```python
# with heat map

c_data.plot(x = 'sepal_length', y = 'petal_length', kind = 'scatter', title = 'Iris Data', c = 'sepal_length')
```

```python
# With Secondary Axis

ax = c_data.plot(kind = 'scatter', x = 'sepal_width', y = 'petal_width', label = 'width', color = 'y')
c_data.plot(kind = 'scatter', x = 'sepal_length', y = 'petal_length', label = 'length', ax = ax, color = 'k')
```

# Data Visualization (Pandas - Hex Plot)

```python
c_data
```

```python
c_data.plot(x = 'sepal_length', y = 'petal_width', gridsize = 20, kind = 'hexbin', C = 'sepal_width')
```

# Data Visualization (Pie)

```python
c_data.iloc[0].plot(kind = 'pie', figsize = (10,5))
```

```python
# Percentage Labels

c_data.iloc[0].plot(kind = 'pie', figsize = (10,5), autopct = '%.2f')
```

# Data Visualization (Scatter Matrix)

```python
# import scatter matrix

from pandas.plotting import scatter_matrix
```

```python
# Scatter Matrix

scatter_matrix(c_data, figsize = (10,5), diagonal = 'kde', color = 'b') # KDE is Kernel Density Estimate
```
