```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Generate Data

```python
# x-axis

x = np.linspace(1,20,20)
x
```

```python
# y-axis

np.random.seed(10)

y = np.random.randint(1,50,20)
y
```

```python
# Use plt (functional - style)

plt.plot(y)
plt.show()
```

```python
# Cumulative

y = y.cumsum()
y
```

```python
# Plot again

plt.plot(x,y)
```

### Change Color and Style

```python
# Color (constant or hex values), Style (:, -., --, -)

plt.plot(x, y, color = 'r', marker = 'P', linestyle = '--')
```

### Sizing, Labels, Title

```python
plt.figure(figsize = (10,5))
plt.plot(x, y, color = '#123434', marker = 'x', linestyle = '--', linewidth = 3)
plt.xlabel('X-Axis', fontsize = 12, fontstyle = 'italic')
plt.ylabel('Y-Axis', fontsize = 12, rotation = 'horizontal')
plt.title('Synthetic Graph', fontsize = 14)
```

### Scatter, Bar, Hist, Box Plot

```python
# Scatter

plt.figure(figsize = (12,5))
plt.scatter(x, y, marker = 's', color = 'g', linewidth = 5)
```

### Bar

# Vertical Bar

b = [10,23,43,5,66]
a = ['a','b','c','d','e']

```python
plt.bar(a, b, width = 0.75, color = 'k')
```

```python
# Horizontal Bar

plt.barh(a, b, height = 0.75)
```

```python
### Histogram

plt.hist(y, rwidth = 0.75, bins = 30)
plt.show()
```

```python
### Histogram

plt.hist(y, orientation = 'horizontal', rwidth = 0.75, bins = 30)
plt.show()
```

```python
# Box Plot

data = [np.random.normal(0, std, 100) for std in range(1,3)]
data
```

```python
plt.boxplot(data, vert = True, patch_artist = True)
plt.show()
```

### Subplots

```python
# Data
x
```

```python
# Data
y
```

```python
y2 = x * y
y2
```

```python
plt.figure(figsize = (11,5))
plt.plot(x, y)
plt.plot(x, y2)
plt.show()
```

```python
# Plot lines separately

plt.figure(figsize = (12,5))
plt.subplot(3,1,2)
plt.plot(x, y, marker = 'o', color = 'r', markersize = 8)

plt.subplot(3,1,1)
plt.plot(x, y2, marker = '*', color = 'k', markersize = 8, linestyle = '--')

plt.subplot(3,1,3)
plt.plot(x, y, marker = '^', color = 'r', markersize = 8)
```

```python
# Nested Plots (Advanced) - OOP way ng matplotlib

fig = plt.figure(figsize = (12,5), dpi = 300)
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.12, 0.6, 0.4, 0.3]) # left, bottom, size, size
ax3 = fig.add_axes([0.55, 0.1, 0.4, 0.3])

ax1.plot(x, y, color = 'r')
ax1.set_xlabel('x-axis')
ax1.set_ylabel('y-axis')
ax1.set_title('Main Graph')

ax2.plot(x, y, color ='g', linestyle = '--')
ax2.set_xlabel('x-axis', fontsize = 8)
ax2.set_ylabel('y-axis', fontsize = 8)
ax2.set_title('Secondary Graph 1')

ax3.plot(x, y, color ='g', linestyle = '--')
ax3.set_xlabel('x-axis', fontsize = 8)
ax3.set_ylabel('y-axis', fontsize = 8)
ax3.set_title('Secondary Graph 2')
```

### xlim, ylim, xticks, yticks

```python
fig, ax = plt.subplots(1,2, figsize = (12,5))

ax[0].plot(x, y)
ax[0].plot(x, y2)
ax[1].plot(x, y2, color = 'k')
```

```python
# with limits

fig, ax = plt.subplots(1,2, figsize = (12,5))

ax[0].plot(x, y)
ax[0].plot(x, y2)
ax[1].plot(x, y2, color = 'k')

# limits

ax[0].set_ylim([0, 15000])
ax[0].set_xlim([0, 50])
ax[0].set_title('Graph 1')

x_limits = np.arange(0, 22, 2)
y_limits = np.arange(0, 10100, 1000)

ax[1].set_xticks(x_limits.tolist())
ax[1].set_yticks(y_limits.tolist())
ax[0].set_title('Graph 2')

plt.show()
```

### Pie Plot

```python
# Data

labels = ['Cat', 'Dog', 'Hamster', 'Rabbit']
sizes = [15, 30, 45, 80]
```

```python
import matplotlib as mpl
mpl.rcParams['font.size'] = 4.0 # global settings

fig, ax = plt.subplots(figsize = (2, 2), dpi = 200)
ax.pie(sizes, labels = labels, autopct = '%.1f', shadow = True, startangle = 90, explode = (0.2, 0.5, 0.3, 0.75))
plt.show()
```
